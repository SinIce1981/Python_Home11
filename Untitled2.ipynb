# Дана функция f(x) = (sin(x)/2 + cos(x)/2) ** 2 * sin(x)

```python
from sympy import *

```

```python
x = Symbol('x')

function = (sin(x)/2 + cos(x)/2) ** 2 * sin(x)
function
```

### Определение корней

```python
solveset(function, x)
```

### Корни функции одного периода

```python
root = solve(function, x)
print(root)

```

```python
for i in root:
    print(i.evalf())
   
```

### Найти интервалы, на которых функция возрастает
(функция возрастает на интервале, на котором значение ее производной больше 0)

```python
derivative = diff(function)
derivative
```

### Построить график 

```python
graf = plot(function, derivative, (x, -pi/4, 2*pi), title = "График функции и ее производной")
```

### Вычислить вершину

```python
for x in derroot:
    y = (sin(x)/2 + cos(x)/2) ** 2 * sin(x)
    print (f'x = {x}, y = {y}')
    print (f'x = {x.evalf()}, y = {y.evalf()}')
    print()
```

### Определить промежутки, на котором f > 0

```python
solve (function > 0,x)
```

### Определить промежутки, на котором f < 0

```python
solve(function < 0,x)
```
